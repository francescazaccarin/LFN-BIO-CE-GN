```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# carichiamo i risultati
centralities = pd.read_csv("centralities.csv")
corr = pd.read_csv("centrality_correlation.csv", index_col=0)

centralities.head()
```

```python


plt.figure(figsize=(6,5))
sns.heatmap(
    corr,
    annot=True,
    cmap="viridis",
    fmt=".2f"
)
plt.title("Correlation between centrality measures")
plt.tight_layout()
plt.show()
```
